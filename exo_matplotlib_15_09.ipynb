```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Tracer un graphe linéaire des points (x, y) où x = [0, 1, 2, 3, 4] et y = [1, 4, 3, 8, 5].

```python
x = [0, 1, 2, 3, 4]
y = [1, 4, 3, 8, 5]

plt.plot(x,y)
plt.show()
```

2. Tracer un histogramme à partir de données aléatoires générées par NumPy.

```python
import numpy as np

a = np.random.randn(500)

plt.hist(a, (20))
plt.show()
```

3. Tracer un diagramme à barres à partir des données suivantes : categories = ['A', 'B', 'C', 'D', 'E'], valeurs = [25, 30, 15, 20, 10]

```python
x =  ['A', 'B', 'C', 'D', 'E']
y = [25, 30, 15, 20, 10]

plt.bar(x,y)
plt.xlabel("catégories")
plt.ylabel("valeurs")
plt.show()
```

4. Tracer un nuage de points à partir de données aléatoires générées par NumPy.

```python
plt.plot(np.random.randint(0, high=100, size=30),'o')
plt.show()
```

5. Tracer une courbe sinusoïdale en utilisant NumPy.

```python
x = np.linspace(0, 10, 100, True)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

6. Tracer deux courbes sur le même graphe à partir des données suivantes : x = np.linspace(0, 10, 100) y1 = np.sin(x) y2 = np.cos(x)

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

7. Tracer un diagramme en secteurs (camembert) à partir des données suivantes : categories = ['A', 'B', 'C', 'D'], valeurs = [30, 25, 15, 20]

```python
y = ['A', 'B', 'C', 'D']
x = [30, 25, 15, 20]

plt.pie(x,labels=y)
plt.show()
```

8. Tracer une courbe polynomiale à partir des données suivantes : x = np.linspace(-5, 5, 100) y = 3 * x3 + 2 * x2 - 8 * x + 5

```python
x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

plt.plot(x, y)
plt.show()
```

9. Tracer un diagramme en barres empilées à partir des données suivantes : categories = ['A', 'B', 'C', 'D'] valeurs1 = [10, 15, 20, 25], valeurs2 = [5, 10, 15, 20]

```python
x =  ['A', 'B', 'C', 'D']
y1 = [10, 15, 20, 25]
y2 = [5, 10, 15, 20]

plt.bar(x,y1)
plt.bar(x,y2)
plt.show()
```

10. Tracer un graphique en boîte (boxplot) à partir des données suivantes :
data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(-3, 2, 100)]
