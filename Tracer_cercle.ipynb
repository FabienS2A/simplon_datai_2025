```python
x = np.linspace(0, 2*np.pi, 1000)



y1 = np.cos(x)
y2 = np.sin(x)


plt.scatter(y1, y2)


plt.xlim([-2,2])
plt.ylim([-2,2])
plt.axis("scaled")

plt.show()
```
