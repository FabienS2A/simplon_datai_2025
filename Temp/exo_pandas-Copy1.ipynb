Voici un façon de créer une DataFrame à partir d'un dictionnaire python:

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "colonne a": [1, 2, 3],
    "colonne b": [6, 7, 8],
    "colonne c": [0, 8, 1]
})

print(df)
```

Rajouter une nouvelle colonne.

Renommer une colonne, inplace = True permet de renommer sur le tableau original sans en créer de nouveau.

```python
df.rename(columns = {'colonne D':'colonne d'}, inplace = True)

df
```

Renommer toues le colonne mettant les initial en majuscule (str.title?)

```python
df.rename(str.title,axis='columns',inplace=True)

df
```

Quelle est la période temporelle sur laquelle s'étendent les données ?

-> On commence par sortir les deux premières colonnes du tableau ou sont contenues les dates
-> dates.apply(pd.to_datetime) permet de changer la colonne 2 en format date. Attention, cette fontion ne permet pas de modifier le tableau d'originne. Pour ça on change le tableau avec "dates =...". On pourrais aussi le spécifier avec "inplace = True" en argument.

```python
dates = df[2:]["Unnamed: 0"]

dates = dates.apply(pd.to_datetime)

dates
```

```python
-> afficher la date max
```

```python
dates.max()
```

```python
-> afficher la date min
```

```python
dates.min()
```

```python
calculer le nombre de jour, en soustrayant le max au min
```

```python
date_range =(dates.max() - dates.min())

date_range.days+1

print("Les donnée s'étalent sur", date_range.days+1, "jours")
```
