Measurer la corr√©lation entre :
two random set


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(0,100,1000)
y = np.random.randint(0,100,1000)


plt.scatter(x,y)
pearson_corr = np.corrcoef(x,y)

print(pearson_corr)
plt.show()
```

bewtween a cosinus and a sinus variables


```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)



pearson_corr = np.corrcoef(y1,y2)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
print(pearson_corr)
plt.show()
```

two gaussian distributed data with different mean and variance (play with the parameters)



```python

x = np.random.normal(loc=0,size=1000)
y = np.random.normal(loc=4,size=1000)



pearson_corr = np.corrcoef(x,y)

print(pearson_corr)
plt.hist(x,(20))
plt.hist(y,(20))
```

```python
x = np.random.normal(loc=0,scale=1,size=1000)
y = np.random.normal(loc=20,scale=3,size=1000)



pearson_corr = np.corrcoef(x,y)

print(pearson_corr)
plt.hist(x,(20))
plt.hist(y,(20))
```
