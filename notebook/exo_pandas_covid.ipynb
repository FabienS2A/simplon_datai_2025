Quelle est la période temporelle sur laquelle s'étendent les données ?

-> On commence par sortir les deux premières colonnes du tableau ou sont contenues les dates
-> dates.apply(pd.to_datetime) permet de changer la colonne 2 en format date. Attention, cette fontion ne permet pas de modifier le tableau d'originne. Pour ça on change le tableau avec "dates =...". On pourrais aussi le spécifier avec "inplace = True" en argument.

```python
dates = df[2:]["Unnamed: 0"]

dates = dates.apply(pd.to_datetime)

dates
```

Tracer l'evolution sur un même graphique des hospitalisations et des réanimations, pour les Femmes et pour les hommes. Discuter les résultats

```python
fig, ax = plt.subplots(figsize=(12, 6))


ax.plot(dates, df[2:]["hosp.1"], color='blue', label='hospitalisa° M')
ax.plot(dates, df[2:]["hosp.2"], color='red', label='hospitalisa° F')
ax.plot(dates, df[2:]["rea.1"], color='cyan', label='reanima° M')
ax.plot(dates, df[2:]["rea.2"], color='magenta', label='reanima° F')


plt.xticks(rotation=45, horizontalalignment='right')  # Rotate xtick labels for better readability !
plt.xlabel("Dates")
plt.ylabel("population")
plt.legend()

plt.show()
```

Tracer sut le même graphique le nombre de décès cumulés pour les hommes et pour les femmes, mais sur une échelle verticale différente/sépare (plt.twinx?)

```python
fig, ax = plt.subplots(figsize=(12, 6))

x = dates

ax.plot(dates, df[2:]["hosp.1"], color='blue', label='hospitalisa° M')
ax.plot(dates, df[2:]["hosp.2"], color='red', label='hospitalisa° F')
ax.plot(dates, df[2:]["rea.1"], color='cyan', label='reanima° M')
ax.plot(dates, df[2:]["rea.2"], color='magenta', label='reanima° F')
ax.legend()
ax.set_ylabel("type d'hosp", color='red')

ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis
ax2.plot(dates, df[2:]["dc.1"], color='orange', label='décé M')
ax2.plot(dates, df[2:]["dc.2"], color='yellow', label='décé F')
ax2.legend()
ax2.set_ylabel("décés cumulé", color='red')


plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

plt.show()
```
