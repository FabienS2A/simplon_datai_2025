```python
conditionement = df["condition_id"]
conditionement 
```

```python
def create_dim_t(y): # créer un tableau dim avec clefs unique
    z = []
    for x in y :
    # check if exists in unique_list or not
        if x not in z:
            z.append(x)
    df = pd.DataFrame(z)
    return df

condition_id = create_dim_t(conditionement)
type(condition_id)
condition_id
condition_id.to_csv("condition_dim.csv")

```

```python
famille = df["famille_id"]
famille
```

```python
famille_id = create_dim_t(famille)
type(famille_id)
famille_id
famille_id.to_csv("famille_dim.csv")
famille_id
```
